```python
# ==========================================================
# GOOGLE COLAB - QUALITY CHECK FOR ALL 3 PROJECT FILES
# Files:
#   1. balanced_1000_per_month.csv
#   2. unique_monthly_capped_1000.csv
#   3. monthly_counts.csv
#
# Best approach:
# Check all 3 once, then use balanced file as main dataset.
# ==========================================================

import pandas as pd


# -----------------------------------
# LOAD FILES
# -----------------------------------
balanced = pd.read_csv("/content/balanced_1000_per_month.csv")
unique_df = pd.read_csv("/content/unique_monthly_capped_1000.csv")
counts = pd.read_csv("/content/monthly_counts.csv")

# ==========================================================
# FUNCTION FOR SEQUENCE DATASET CHECK
# ==========================================================
def inspect_seq_df(df, name):

    print("\n" + "="*60)
    print(name.upper())
    print("="*60)

    print("Shape:", df.shape)

    print("\nColumns:")
    print(df.columns.tolist())

    print("\nMissing Values:")
    print(df.isna().sum())

    print("\nDuplicate EPI_ISL:",
          df["EPI_ISL"].duplicated().sum())

    print("Duplicate Sequence:",
          df["Sequence"].duplicated().sum())

    print("\nMonth Counts:")
    print(df["YearMonth"].value_counts().sort_index())

    # Sequence length
    df["SeqLen"] = df["Sequence"].astype(str).str.len()

    print("\nSequence Length Distribution:")
    print(df["SeqLen"].value_counts().sort_index())

    # Character check
    valid_aa = set("ACDEFGHIKLMNPQRSTVWY")
    chars = set("".join(df["Sequence"].astype(str)))
    invalid = chars - valid_aa

    print("\nUnique Characters Found:", sorted(chars))
    print("Invalid Characters:", sorted(invalid))

    print("\nTop 10 Countries:")
    print(df["Country"].value_counts().head(10))

    print("\nRAM Usage (MB):",
          round(df.memory_usage(deep=True).sum()/1024**2,2))

    print("\nFirst 3 Rows:")
    print(df.head(3))

# ==========================================================
# CHECK DATASET 1
# ==========================================================
inspect_seq_df(balanced, "balanced_1000_per_month.csv")

# ==========================================================
# CHECK DATASET 2
# ==========================================================
inspect_seq_df(unique_df, "unique_monthly_capped_1000.csv")

# ==========================================================
# CHECK DATASET 3
# ==========================================================
print("\n" + "="*60)
print("MONTHLY_COUNTS.CSV")
print("="*60)

print("Shape:", counts.shape)

print("\nColumns:")
print(counts.columns.tolist())

print("\nMissing Values:")
print(counts.isna().sum())

print("\nFirst 10 Rows:")
print(counts.head(10))

print("\nLast 10 Rows:")
print(counts.tail(10))

print("\nTotal Sequences Across All Months:")
print(counts["Count"].sum())

print("\nPeak Month:")
peak = counts.loc[counts["Count"].idxmax()]
print(peak)

print("\nLowest Month:")
low = counts.loc[counts["Count"].idxmin()]
print(low)

print("\n========== ALL FILES VERIFIED ==========")
```

```python
# ==========================================================
# GOOGLE COLAB - FINAL CLEAN DATASET FOR MODELING
# Input : balanced_1000_per_month.csv
# Output: balanced_clean_1274.csv
#
# Rules:
# 1. Remove sequences containing X or *
# 2. Keep only length = 1274
# 3. Save final ML-ready dataset
# ==========================================================

import pandas as pd



# -----------------------------------
# Load dataset
# -----------------------------------
df = pd.read_csv("/content/balanced_1000_per_month.csv")

print("Original Shape:", df.shape)

# -----------------------------------
# Ensure sequence string
# -----------------------------------
df["Sequence"] = df["Sequence"].astype(str)

# -----------------------------------
# Sequence Length
# -----------------------------------
df["SeqLen"] = df["Sequence"].str.len()

print("\nOriginal Length Counts:")
print(df["SeqLen"].value_counts().sort_index())

# -----------------------------------
# Remove X and *
# -----------------------------------
mask_valid = ~df["Sequence"].str.contains("X|\*", regex=True)

removed_invalid = (~mask_valid).sum()

df = df[mask_valid]

print("\nRemoved Invalid (X or *):", removed_invalid)

# -----------------------------------
# Keep only length = 1274
# -----------------------------------
df = df[df["SeqLen"] == 1274]

print("\nAfter Length=1274 Filter:", df.shape)

# -----------------------------------
# Reset index
# -----------------------------------
df = df.reset_index(drop=True)

# -----------------------------------
# Final Checks
# -----------------------------------
print("\nMonth Counts:")
print(df["YearMonth"].value_counts().sort_index())

print("\nCountries:", df["Country"].nunique())

print("\nFinal Length Check:")
print(df["Sequence"].str.len().value_counts())

# -----------------------------------
# Drop helper column
# -----------------------------------
df = df.drop(columns=["SeqLen"])

# -----------------------------------
# Save file
# -----------------------------------
df.to_csv("balanced_clean_1274.csv", index=False)

files.download("balanced_clean_1274.csv")

print("\n========== DONE ==========")
print("Saved: balanced_clean_1274.csv")
```
